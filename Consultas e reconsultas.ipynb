```python
import mysql.connector
import decimal
import pprint
import csv
```

```python
con = mysql.connector.connect(host="#######", 
                              user="#######", passwd="#######", db="#######")
cursor = con.cursor(buffered=True)
cursor2 = con.cursor(buffered=True)
cursor3 = con.cursor(buffered=True)
```

```python
relatorio = {}
pretendentes = []
#COLOCAR AQUI O MES E ANO CORRESPONDENTE
mes = 12
ano = 2021
query = "select p.id, p.nome, pp.pretendente_id, pp.valor, pp.payment_at "
query += "from pretendente_produto pp "
query += "inner join pretendente pr on pr.id = pp.pretendente_id "
query += "inner join solicitacao s on pr.solicitacao_id = s.id "
query += "inner join pessoa p on p.id = s.cliente_id "
query += "where month(pp.payment_at) = "+str(mes)+" and year(pp.payment_at) = "+str(ano)+" "
query += "group by pretendente_id order by p.nome"

cursor.execute(query)
for resultado in cursor:
    if resultado[0] not in relatorio:
        relatorio[resultado[0]] = {}
        relatorio[resultado[0]]['nome_cliente'] = resultado[1]
        relatorio[resultado[0]]['consultas'] = 0
        relatorio[resultado[0]]['reconsultas'] = 0
        relatorio[resultado[0]]['consultas_valor'] = 0
        relatorio[resultado[0]]['reconsultas_valor'] = 0
    query = "select count(id) from pretendente_produto where pretendente_id = " + str(resultado[2]) + ""
    cursor2.execute(query)
    for resultado2 in cursor2:
        if resultado2[0] == 1:
            relatorio[resultado[0]]['consultas'] += 1
            relatorio[resultado[0]]['consultas_valor'] += resultado[3]
        else:
            query = "select month(payment_at) from pretendente_produto where pretendente_id = " + str(resultado[2]) + " and month(payment_at) <= "+str(mes)+" and year(payment_at) = "+str(ano)+" and payment_at is not null"
            cursor3.execute(query)
            quantidade_no_mes = 0
            quantidade_fora_do_mes = 0
            for resultado3 in cursor3:
                print(str(resultado[2]) + " " + str(resultado3[0]))
                if resultado3[0] == mes:
                    quantidade_no_mes += 1
                else:
                    quantidade_fora_do_mes += 1
            print('quantidade no mes: ', quantidade_no_mes)
            print('quantidade fora do mes: ', quantidade_fora_do_mes)
            if quantidade_fora_do_mes > 0:
                relatorio[resultado[0]]['reconsultas'] += quantidade_no_mes
                relatorio[resultado[0]]['reconsultas_valor'] += resultado[3]*quantidade_no_mes  
            elif quantidade_no_mes > 0:
                relatorio[resultado[0]]['consultas'] += 1
                relatorio[resultado[0]]['consultas_valor'] += resultado[3]
                relatorio[resultado[0]]['reconsultas'] += quantidade_no_mes-1
                relatorio[resultado[0]]['reconsultas_valor'] += resultado[3]*(quantidade_no_mes-1)
            print('consultas ' + str(relatorio[resultado[0]]['consultas']))
            print('reconsultas ' + str(relatorio[resultado[0]]['reconsultas']))
            print('valor somado: ' + str(resultado[3]))
            print('valor reconsulta: ' + str(relatorio[resultado[0]]['reconsultas_valor']))
    print("Dia " + str(resultado[4]) + " analisado")
    print('-------------------')
```

```python
relatorio1 = relatorio
csv_columns = ['ID', 'CLIENTE', 'QUANTIDADE CONSULTAS', 'VALOR CONSULTAS', 
               'QUANTIDADE RECONSULTAS', 'VALOR RECONSULTAS',
              'QUANTIDADE TOTAL', 'VALOR TOTAL']

with open('relatorio_consultas_reconsultas.csv', 'w', newline='', encoding='utf-8') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';')
    spamwriter.writerow(csv_columns)
    for key in relatorio1:
        listaRow = list()
        listaRow.append(key)
        listaRow.append(relatorio1[key]['nome_cliente'])
        listaRow.append(relatorio1[key]['consultas'])
        listaRow.append(str(float(relatorio1[key]['consultas_valor'])).replace('.',','))
        listaRow.append(relatorio1[key]['reconsultas'])
        listaRow.append(str(float(relatorio1[key]['reconsultas_valor'])).replace('.',','))
        listaRow.append(relatorio1[key]['consultas']+relatorio1[key]['reconsultas'])
        listaRow.append(str(float(relatorio1[key]['consultas_valor']+relatorio1[key]['reconsultas_valor'])).replace('.',','))

        spamwriter.writerow(listaRow)
        
print('Script executado com sucesso!')
```
